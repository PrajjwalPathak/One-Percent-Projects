### MODEL FOR PREDICTION

```python
from xgboost import XGBRegressor
pd.options.mode.chained_assignment = None
```

```python
fd=pd.read_csv("csvDumps/Intern_FutureDays.csv", parse_dates=['Day'], dayfirst=True)
fd.set_index("Day",inplace=True)
fd
```

#### Prediction for INTERN

```python
df = pd.read_csv('csvdumps/CONSOLE_DIGITAL_MARKETING_PLATFORM_INTERN_APPLE.csv')
df=df.iloc[:,0:1]
df['Day'] = df.index
df.head()
```

```python
test_df = df
train_df = df
X_train, y_train = train_df[['Day']], train_df[['INTERN']]
X_test, y_test = test_df[['Day']], test_df[['INTERN']]
```

```python
reg = XGBRegressor(n_estimators=200,learning_rate=0.01)
reg.fit(X_train, 
        y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        eval_metric='mae')
```

```python
validations = reg.predict(X_test)
test_df['validations'] = validations

test_df=test_df.iloc[:-1,:]
test_df.head()
```

```python
plt.rcParams.update({'figure.figsize': (15, 5), 'figure.dpi':300})
fig, ax = plt.subplots()
sns.lineplot(data=test_df, x=test_df.index, y='INTERN', label ='Original', color='Green')
sns.lineplot(data=test_df, x=test_df.index, y='validations', label ='Predicted', color='Red')
plt.grid(linestyle='-', linewidth=0.3)
plt.xlabel('Day')
plt.ylabel('INTERN')
plt.title('Graph for Daily View Count for the platform')
plt.legend()
ax.tick_params(axis='x', rotation=90)
```

```python
fd['Intern']= test_df['validations'].to_numpy()
fd['Intern']= test_df['validations'].values
fd.head()
```

```python
train_df=train_df.iloc[:-1,:]

plt.rcParams.update({'figure.figsize': (15, 5), 'figure.dpi':300})
fig, ax = plt.subplots(1,2)
sns.lineplot(data=train_df, x=train_df.index, y='INTERN', label ='Original', color='Green',ax=ax[0])
#sns.lineplot(data=mtrain_df, x=mtrain_df.index, y='Facebook', label ='Original', color='Red')
sns.lineplot(data=fd, x=fd.index, y='Intern', label ='Future Predicted', color='Red',ax=ax[1])

plt.legend()
```
