```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv('CONSOLE_DIGITAL_MARKETING_PLATFORM_APP_APPLE.csv')
data
```

## The Bar Graph below shows the total view count for all the apps and tutorials on all the platforms
- If the total view count for any platform is less, then the all the apps and tutorials should be promoted on that platform.

```python
x = []
y = []
for slug in data:
    x.append(slug)
x = x[1:]
for slug in x:
    y.append(int(sum(data[slug])))
plt.figure(figsize=(20,10))
plt.title('Bar Graph showing total view count for all the apps and tutorials on all the platforms', fontsize=20)
plt.xlabel('Platforms', fontsize=15)
plt.ylabel('View Count', fontsize=15)
plt.bar(x, y, color='g')
plt.grid(axis = 'y')
plt.show()
```

- According to the graph above, INSTAGRAM, LINKEDIN, REDDIT, SNAPCHAT, YOUTUBE, TIKTOK, EMAIL and OTHER are the platforms that have very low total view count. All the apps and tutorials must be promoted on these platforms.

## The Plot Graphs below show the view count of each app or each tutorial by the period of time forindividual platforms
- If the view count of an app or a tutorial has a low total view count, then it needs to be promoted on that particular platform.

```python
y = data['APP']
promotions = []
plt.figure(figsize=(20, 125))
for i in range(1,12):
    promotion = []
    plt.subplot(12, 1, i)
    plt.barh(y, data[x[i-1]])
    counter = 0
    for count in data[x[i-1]]:
        if count <= 10:
            promotion.append(y[counter])
        counter = counter + 1
    promotions.append(promotion)
    plt.title(x[i-1], fontsize=15)
    plt.xlabel("View Count", fontsize=15)
    plt.ylabel("Platforms", fontsize=15)
```

### Conclusions - These are the apps and tutorials that must be updated

```python
for i in range(1,12):
    print(x[i-1] + ' - ', end ="")
    for j in promotions[i-1]:
        print(j + ', ', end =" ")  
    print('\n')
```
