```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv('CONSOLE_DIGITAL_MARKETING_APP_DATE_GOOGLE.csv')
data
```

## The Bar Graph below shows the view count of all the apps and tutorials
- If the view count of an app or a tutorial is low then it should be promoted or should be updated in order to increase its view count

```python
x = []
y = []
for slug in data:
    x.append(slug)
x = x[1:]
for slug in x:
    y.append(int(sum(data[slug])))
plt.figure(figsize=(20,10))
plt.title('Bar Graph showing view count for all the apps and tutorials', fontsize=20)
plt.xlabel('View Count', fontsize=15)
plt.ylabel('Apps and Tutorials', fontsize=15)
plt.barh(x,y)
plt.grid(axis = 'x')
plt.show()
```

- According to the graph above, TRAVEL_DIARIES, RUBY_TUTORIAL, RUBY_COMPILER, PHP_COMPILER, KOTLIN_TUTORIAL, KOTLIN_COMPILER, DART_TUTORIAL and DART_COMPILER have the lowest view count among all. They need to be updated.

## The Plot Graphs below show the view count of each app or each tutorial by the period of time
- If the view count of an app or a tutorial has constant spikes then it shows that it was used constantly i.e., users were visiting the page constantly.
- If there was a sudden spike and then no change in the view count then it shows that the page was not used frequently.
- Flat line shows that the app or the tutorial was not used for a long time.

```python
count = 0
fig, axs = plt.subplots(6, 6, figsize=(20,15), sharex=True, sharey=True) 
for i in range(0,6):
    for j in range(0,6):
        axs[i, j].plot(data[x[count]])
        axs[i, j].set_title(x[count])
        count=count+1
fig.supxlabel('Time Period (Dates)', fontsize=15)
fig.supylabel('View Count for Apps and Tutorials', fontsize=15)
```

- According to the graphs above, TRAVEL_DIARIES, RUBY_TUTORIAL, RUBY_COMPILER, PHP_COMPILER, KOTLIN_TUTORIAL, KOTLIN_COMPILER, DART_TUTORIAL and DART_COMPILER have the lowest activity. They need to be promoted or updated.
