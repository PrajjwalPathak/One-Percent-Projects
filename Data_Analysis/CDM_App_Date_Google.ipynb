```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv('CONSOLE_DIGITAL_MARKETING_APP_DATE_GOOGLE.csv')
data
```

```python
x = []
y = []
for slug in data:
    x.append(slug)
for slug in x[1:]:
    y.append(int(sum(data[slug])))
plt.figure(figsize=(20,10))
plt.title('Test')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.barh(x[1:],y)
plt.grid(axis = 'x')
plt.show()
```
