```python
import pandas as pd
import numpy
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
```

### Understanding the dataset

```python
a=pd.read_csv("csvDumps/CONSOLE_DIGITAL_MARKETING_PLATFORM_DATE_GOOGLE.csv")
a["INDEX"] = [i for i  in range(0,320)]
a.to_csv("CONSOLE_DIGITAL_MARKETING_PLATFORM_INTERN_GOOGLE.csv", index=False)
a
```

```python
label=["FACEBOOK","INSTAGRAM","TWITTER","QUORA","LINKEDIN","REDDIT","SNAPCHAT","YOUTUBE","TIKTOK","EMAIL","OTHER"]
```

```python
a.shape
```

```python
total_views=[int(sum(a['FACEBOOK'])),int(sum(a['INSTAGRAM'])),int(sum(a['TWITTER'])),int(sum(a['QUORA'])),int(sum(a['LINKEDIN'])),int(sum(a['REDDIT'])),int(sum(a['SNAPCHAT'])),int(sum(a['YOUTUBE'])),int(sum(a['TIKTOK'])),int(sum(a['EMAIL'])),int(sum(a['OTHER']))]
total_views
```

### Plotting every view of each platform against date

This section plots every view of each platform against the respective date.

As we understand, while Facebook, Instagram, Snapchat, LinkedIn and Reddit are well-used platform, the rest lack the required public engagement.

The well-viewed platforms are identified from the graphs, where in all 320 days, the view is high. Similarly, the less viewed platforms have 320 rows with nearly 0 to less number of views, or even a high view on only 2-3 days max.


```python
plt.figure(figsize=(20,10))
sns.scatterplot(y=a['FACEBOOK'],x=a['INDEX'])
sns.scatterplot(y=a['INSTAGRAM'],x=a['INDEX'])
sns.scatterplot(y=a['TWITTER'],x=a['INDEX'])
sns.scatterplot(y=a['QUORA'],x=a['INDEX'])
sns.scatterplot(y=a['LINKEDIN'],x=a['INDEX'])
sns.scatterplot(y=a['REDDIT'],x=a['INDEX'])
sns.scatterplot(y=a['SNAPCHAT'],x=a['INDEX'])
sns.scatterplot(y=a['YOUTUBE'],x=a['INDEX'])
sns.scatterplot(y=a['TIKTOK'],x=a['INDEX'])
sns.scatterplot(y=a['EMAIL'],x=a['INDEX'])
sns.scatterplot(y=a['OTHER'],x=a['INDEX'])
plt.legend(label)
plt.figure(figsize=(20,20))
```

### Plotting of sum of views for each platform for an overall analysis

```python
plt.pie(total_views,labels=label,autopct='%.0f%%')
plt.title("PIECHART DEPICTING THE SUM OF ALL VISITS IN EACH PLATFORM")
plt.show()
```

```python
label_short=["FB","IG","TW","QRA","LN","RDT","SNP","UT","TT","EM","O"]
sns.barplot(x=label_short, y=total_views)
plt.title("BARPLOT DEPICTING THE SUM OF ALL VIEWS FOR EACH PLATFORM")
plt.show()
```

### Individual Analysis of each platform

As mentioned, from the graph it is clear that the well-viewed apps include - Facebook, Instagram, Linkedin, reddit, snapchat, and likewise the rest fall under the less-viewed apps. For example, "Email" and "Tiktok" have almost 0 to no views at all in all the 320 days. 

Youtube and Quora have high views at the end of the graph. This means that only on certain end days, it has been used well. We want apps to be used daily. Hence a little focus is needed here as well.


```python
fig, axs = plt.subplots(4, 3)
axs[0, 1].plot(a['FACEBOOK'])
axs[0, 1].set_title("FACEBOOK")
axs[0, 2].plot(a['INSTAGRAM'])
axs[0, 2].set_title("INSTAGRAM")
axs[1, 0].plot(a['TWITTER'])
axs[1, 0].set_title("TWITTER")
axs[1, 1].plot(a['QUORA'])
axs[1, 1].set_title("QUORA")
axs[1, 2].plot(a['LINKEDIN'])
axs[1, 2].set_title("LINKEDIN")
axs[2, 0].plot(a['REDDIT'])
axs[2, 0].set_title("REDDIT")
axs[2, 1].plot(a['SNAPCHAT'])
axs[2, 1].set_title("SNAPCHAT")
axs[2, 2].plot(a['YOUTUBE'])
axs[2, 2].set_title("YOUTUBE")
axs[3, 0].plot(a['TIKTOK'])
axs[3, 0].set_title("TIKTOK")
axs[3, 1].plot(a['EMAIL'])
axs[3, 1].set_title("EMAIL")
axs[0, 0].plot(a['OTHER'])
axs[0, 0].set_title("OTHERS")
plt.show()
```
