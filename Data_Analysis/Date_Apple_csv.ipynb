```python
import pandas as pd
import numpy
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

### Understanding the dataset

```python
a=pd.read_csv("csvDumps/CONSOLE_DIGITAL_MARKETING_PLATFORM_DATE_APPLE.csv")
a["INDEX"] = [i for i  in range(0,320)]
a.to_csv("CONSOLE_DIGITAL_MARKETING_PLATFORM_DATE_APPLE.csv", index=False)
a
```

```python
label=["FACEBOOK","INSTAGRAM","TWITTER","QUORA","LINKEDIN","REDDIT","SNAPCHAT","YOUTUBE","TIKTOK","EMAIL","OTHER"]
```

```python
a.shape
```

```python
total_views=[int(sum(a['FACEBOOK'])),int(sum(a['INSTAGRAM'])),int(sum(a['TWITTER'])),int(sum(a['QUORA'])),int(sum(a['LINKEDIN'])),int(sum(a['REDDIT'])),int(sum(a['SNAPCHAT'])),int(sum(a['YOUTUBE'])),int(sum(a['TIKTOK'])),int(sum(a['EMAIL'])),int(sum(a['OTHER']))]
total_views
```

### Plotting every view of each platform against date

This section plots every view of each app on a single graph.

As we understand from the graph, the engagement of the apps is very less. We have 320 days available in the datasets. For each data, only the ending few days have some views.

Out of them, Facebook, Instagram, Twitter, Quora, Linkedin have better views in the last few days. Whereas, the rest have very less views, especially emails and tiktok.

Observe the graphs for better understanding

```python
plt.figure(figsize=(20,10))
plt.plot(a["INDEX"],a["FACEBOOK"],"-.")
plt.plot(a["INDEX"],a["INSTAGRAM"],":")
plt.plot(a["INDEX"],a["TWITTER"],"--")
plt.plot(a["INDEX"],a["QUORA"])
plt.plot(a["INDEX"],a["LINKEDIN"])
plt.plot(a["INDEX"],a["REDDIT"])
plt.plot(a["INDEX"],a["SNAPCHAT"],":")
plt.plot(a["INDEX"],a["YOUTUBE"],"--")
plt.plot(a["INDEX"],a["TIKTOK"])
plt.plot(a["INDEX"],a["EMAIL"])
plt.plot(a["INDEX"],a["OTHER"])
plt.legend(label)
plt.show()
```

### Zooming the graph towards the end of the date timeline for better understanding

```python
plt.figure(figsize=(20,10))
plt.plot(a["INDEX"],a["FACEBOOK"],"-.")
plt.plot(a["INDEX"],a["INSTAGRAM"],":")
plt.plot(a["INDEX"],a["TWITTER"],"--")
plt.plot(a["INDEX"],a["QUORA"])
plt.plot(a["INDEX"],a["LINKEDIN"])
plt.plot(a["INDEX"],a["REDDIT"])
plt.plot(a["INDEX"],a["SNAPCHAT"],":")
plt.plot(a["INDEX"],a["YOUTUBE"],"--")
plt.plot(a["INDEX"],a["TIKTOK"])
plt.plot(a["INDEX"],a["EMAIL"])
plt.plot(a["INDEX"],a["OTHER"])
plt.axis([245, 0, 0, 0])
plt.legend(label)
ax=plt.gca()
ax.set_xlim([230,325])
ax.set_ylim([0,100])
plt.show()
```

### Plotting of sum of views for each platform for an overall analysis

```python
plt.pie(total_views,labels=label,autopct='%.0f%%')
plt.title("PIECHART DEPICTING THE SUM OF ALL VISITS IN EACH PLATFORM")
plt.show()
```

```python
label_short=["FB","IG","TW","QRA","LN","RDT","SNP","UT","TT","EM","O"]
sns.barplot(x=label_short, y=total_views)
plt.title("BARPLOT DEPICTING THE SUM OF ALL VIEWS FOR EACH PLATFORM")
plt.show()
```

Facebook, Instagram, Linkedin, Quora and Twitter have better views comparatively.
The rest have lesser views.
