```python
import pandas as pd
import numpy
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
```

### Understanding the dataset

```python
a=pd.read_csv("CONSOLE_DIGITAL_MARKETING_PLATFORM_INTERN_GOOGLE.csv")
a["INDEX"] = [i for i  in range(0,46)]
a.to_csv("CONSOLE_DIGITAL_MARKETING_PLATFORM_INTERN_GOOGLE.csv", index=False)
a
```

```python
a.shape
```

```python
total_views=[int(sum(a['INTERN'])),int(sum(a['FACEBOOK'])),int(sum(a['INSTAGRAM'])),int(sum(a['TWITTER'])),int(sum(a['QUORA'])),int(sum(a['LINKEDIN'])),int(sum(a['REDDIT'])),int(sum(a['SNAPCHAT'])),int(sum(a['YOUTUBE'])),int(sum(a['TIKTOK'])),int(sum(a['EMAIL'])),int(sum(a['OTHER']))]
total_views
```

```python
label=["INTERN","FACEBOOK","INSTAGRAM","TWITTER","QUORA","LINKEDIN","REDDIT","SNAPCHAT","YOUTUBE","TIKTOK","EMAIL","OTHER"]
```

```python
rows=[]
for i in range(0,12):
    r=a.iloc[:,i]
    r=list(r)
    rows.append(r)
rows
```

### Plotting graphs on the sum of all views of each platform

I have added 2 sections here. One is the sum of views for each platform, and the other half is plotting every view of each platform.
This helps in understanding the growth. There might be a platform with only 1 single viewer that has a high view. While another platform has many viewers but the viewcount is less. 
This comparison helps us understand the psychology of the users as well as the impact of a platform.

```python
fig, ax = plt.subplots(4, 3)
sns.countplot(x=a['INTERN'],data=a,ax=ax[0,0]).set_title("INTERN")
sns.countplot(x=a['FACEBOOK'],data=a, ax=ax[0,1]).set_title("FACEBOOK")
sns.countplot(x=a['INSTAGRAM'],data=a ,ax=ax[0,2]).set_title("INSTAGRAM")
sns.countplot(x=a['TWITTER'],data=a,ax=ax[1,0]).set_title("TWITTER")
sns.countplot(x=a['QUORA'],data=a,ax=ax[1,1]).set_title("QUORA")
sns.countplot(x=a['LINKEDIN'],data=a,ax=ax[1,2]).set_title("LINKEDIN")
sns.countplot(x=a['REDDIT'],data=a,ax=ax[2,0]).set_title("REDDIT")
sns.countplot(x=a['SNAPCHAT'],data=a,ax=ax[2,1]).set_title("SNAPCHAT")
sns.countplot(x=a['YOUTUBE'],data=a,ax=ax[2,2]).set_title("YOUTUBE")
sns.countplot(x=a['TIKTOK'], data=a,ax=ax[3,0]).set_title("TIKTOK")
sns.countplot(x=a['EMAIL'],data=a,ax=ax[3,1]).set_title("EMAIL")
sns.countplot(x=a['OTHER'],data=a,ax=ax[3,2]).set_title("OTHER")
plt.title("COUNTPLOT OF EACH PLATFORM")
```

```python
plt.pie(total_views,labels=label,autopct='%.0f%%')
plt.title("PIECHART DEPICTING THE SUM OF ALL VISITS IN EACH PLATFORM")
plt.show()
```

```python
label_short=["INT","FB","IG","TW","QRA","LN","RDT","SNP","UT","TT","EM","O"]
sns.barplot(x=label_short, y=total_views)
plt.title("BARPLOT DEPICTING THE SUM OF ALL VIEWS FOR EACH PLATFORM")
plt.show()
```

```python
sns.scatterplot(x=label_short, y=total_views,hue=label)
plt.title("SCATTERPLOT OF SUM OF ALL VIEWS FOR EACH PLATFORM")
```

### Analyzing Facebook, Instagram, LinkedIn and Reddit for having good views
Facebook has a good number of views in some cases that is spreaded in the scatterplot. Hence,it means that quite a few people use facebook on a good amount. However it still holds quite a lot of users that do not use facebook that well. So perhaps this could mean a lack of features or advertizing new features ( to especially the younger generation).

Instagram again, has some users that view it. However, that is very less. Hence, more publicity and getting it introduced to the older generation might help.

Linkedin is a platform mostly used for jobs. Hence the number of good users is less comparatively than facebook but again, more than instagram.

Reddit has a good number of users who give good views. The scatter plot is spreaded (much like fb), but again quite of  the, don't give in much views.



```python
fig, ax = plt.subplots(4, 2)

sns.distplot(a['FACEBOOK'], ax=ax[0,0]).set_title("FACEBOOK")
sns.scatterplot(y=a['FACEBOOK'],x=a['INDEX'], ax=ax[0,1]).set_title("FACEBOOK")
sns.distplot(a['INSTAGRAM'],ax=ax[1,0]).set_title("INSTAGRAM")
sns.scatterplot(y=a['INSTAGRAM'],x=a['INDEX'] ,ax=ax[1,1]).set_title("INSTAGRAM")
sns.distplot(a['LINKEDIN'],ax=ax[2,0]).set_title("LINKEDIN")
sns.scatterplot(y=a['LINKEDIN'],x=a['INDEX'] ,ax=ax[2,1]).set_title("LINKEDIN")
sns.distplot(a['REDDIT'],ax=ax[3,0]).set_title("REDDIT")
sns.scatterplot(y=a['REDDIT'],x=a['INDEX'] ,ax=ax[3,1]).set_title("REDDIT")
```
